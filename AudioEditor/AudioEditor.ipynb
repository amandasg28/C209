```python
import sounddevice as sd 
import numpy as np
from scipy.io import wavfile

flag = True

print("Aviso: É necessário que o áudio a ser editado esteja na mesma pasta que o código")

while flag:
    print("1) Entrar com o nome do áudio")
    print("2) Reproduzir o áudio")
    print("3) Cortar o áudio")
    print("4) Modificar a frequência do áudio")
    print("5) Mudar o volume do áudio")
    print("6) Salvar o audio")
    a = int(input("Escolha uma das opções:"))
    if a==1:
        audioName = input("Nome do arquivo:\n")
        fs, data = wavfile.read(audioName+'.wav')
    elif a==2:
        sd.play(data, fs)
    elif a==3:
        print("1) Cortar pelo final")
        print("2) Cortar pelo início")
        op=int(input("Opção:\n"))
        seconds=int(input("Segundos:\n"))
        if op == 1:  
            fs = int(fs)
            data = data[:(-1)*(fs*seconds), :]
        elif op == 2:
            fs = int(fs)
            data = data[fs*seconds:, :]
    elif a==4:
        escalar =  float(input("Valor da frequencia: Maior que 0 aumenta, entre 0 e 1 diminui\n"))
        fs = fs*escalar
    elif a==5:
        escalar =  float(input("Valor do volume: Maior que 0 aumenta, entre 0 e 1 diminui\n"))
        data = escalar*data
        
    elif a==6:
        new_name = input("Novo nome do audio:\n")
        wavfile.write(new_name+".wav", fs, data)
        print('Áudio salvo com sucesso na pasta onde se encontra o código!!')
        flag = False
    else :
        print("Escolha Inválida\n")
```
